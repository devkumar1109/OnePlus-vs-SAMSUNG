# Data Gathering

```python
import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# Setup the ChromeDriver path using webdriver_manager
service = Service('C:/Users/user/Desktop/chromedriver.exe')

# Initialize the Chrome WebDriver
driver = webdriver.Chrome(service=service)

# URL of the Flipkart OnePlus phones page
url = 'https://www.flipkart.com/mobiles/~samsung-galaxy/pr?sid=tyy%2C4io&page=1'
# Open the URL with the WebDriver
driver.get(url)
time.sleep(5)  # Wait for the page to load

# Get the page source and parse it with BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')
soup.prettify()
products = soup.find_all('div', class_='tUxRFH')
model = []
ratings_reviews= []
discounted_price= []
discount= []
rating = []
camera = []
battery = []
ram_rom = []
display = []
original_price = []
product_image = []
for i in products:
    
    
        try:
            model.append(i.find('div', class_= 'KzDlHZ').text.strip())
        except:
            model.append(np.nan)
        try:
            ratings_reviews.append(i.find('span', class_="Wphh3N").text.strip())
        except:
            ratings_reviews.append(np.nan)
        try:   
            discounted_price.append(i.find('div', class_="Nx9bqj _4b5DiR").text.strip())
        except:
            discounted_price.append(np.nan)
        try:   
            original_price.append(i.find('div', class_="yRaY8j ZYYwLA").text.strip())
        except:
            original_price.append(np.nan)
        try:
            discount.append(i.find('div', class_= 'UkUFwK').text.strip())
        except:
            discount.append(np.nan)
        try:
            rating.append(i.find('div', class_="XQDdHH").text.strip())
        except:
            rating.append(np.nan)
        try:
            ram_rom.append(i.find_all('li', class_="J+igdf")[0].text.strip())
        except:
            ram_rom.append(np.nan)
        try:
            display.append(i.find_all('li', class_ = 'J+igdf')[1].text.strip())
        except:
            display.append(np.nan)
        try:
            camera.append(i.find_all('li', class_="J+igdf")[2].text.strip())
        except:
            camera.append(np.nan)
        try:
            battery.append(i.find_all('li', class_="J+igdf")[3].text.strip())
        except:
            battery.append(np.nan)
        try:
            product_image.append(i.find('img', class_="DByuf4")['src'])
        except:
            product_image.append(np.nan)
d = {'Model': model, 'Number of rating and reviews': ratings_reviews, 'Discounted_Price': discounted_price, 'Discount': discount, 'Rating': rating, 'ram_rom': ram_rom, 'Battery': battery, 'Display': display, 'Camera': camera, 'original_price': original_price, 'product_image': product_image}
dff = pd.DataFrame(d)
    
```

# Data Cleaning

```python
samsung = pd.read_csv('C:/Users/user/Downloads/samsung_flipkart_phones.csv')
```

```python
samsung.sample(3)
```

```python
samsung.duplicated().sum()
```

```python
samsung.drop_duplicates(inplace = True)
```

```python
samsung.info()
```

```python
samsung['Discount'] = samsung['Discount'].fillna('0% off')
```

```python
samsung
```

```python
for i in range(samsung.shape[0]):
    if pd.isna(samsung['original_price'].iloc[i]):
        samsung['original_price'].iloc[i] = samsung['Discounted_Price'].iloc[i]
```

```python
samsung
```

```python
samsung.dropna(subset=['Discounted_Price'], inplace = True)
```

```python
samsung
```

```python
samsung['Discounted_Price'] = samsung['Discounted_Price'].str.replace('₹', '').str.replace(',', '').astype(int)
```

```python
samsung['original_price'] = samsung['original_price'].str.replace('₹', '').str.replace(',', '').astype(int)
```

```python
samsung['Discount'] = samsung['Discount'].apply(lambda x: x.split('%')[0]).astype(int)
```

```python
samsung
```

```python
samsung.info()
```

```python
samsung.dropna(inplace = True)
```

```python
samsung['No_of_Ratings'] = samsung['Number of rating and reviews'].apply(lambda x: x.split(' ')[0])
```

```python
samsung['No_of_Reviews'] = samsung['Number of rating and reviews'].apply(lambda x: x.split('&')[1])
```

```python
samsung['No_of_Reviews'] = samsung['No_of_Reviews'].apply(lambda x: x.split(' ')[0])
```

```python
samsung['No_of_Ratings'] = samsung['No_of_Ratings'].str.replace(',', '').astype(int)
samsung['No_of_Reviews'] = samsung['No_of_Reviews'].str.replace(',', '').astype(int)
```

```python
samsung
```

```python
samsung = samsung[samsung['Battery'].str.contains('Battery')]
```

```python
samsung
```

```python
samsung['Battery_in_mAh'] = samsung['Battery'].apply(lambda x: x.split(' ')[0]).astype(int)
```

```python
samsung
```

```python
samsung = samsung[samsung['Display'].str.contains('Display')]
```

```python
samsung['Display_size_cm'] = samsung['Display'].apply(lambda x: x.split('(')[0].split(' ')[0]).astype(float)
```

```python
samsung
```

```python
samsung['Camera'].unique()
```

```python
samsung[~samsung['Camera'].str.contains('Camera')]
```

```python
samsung['Camera'] = samsung['Camera'].apply(lambda x: x.split(' ')[0])
```

```python
samsung
```

```python
samsung['model'] = samsung['Model'].apply(lambda x: x.split('(')[0])
```

```python
samsung
```

```python
samsung['Color'] = samsung['Model'].apply(lambda x: x.split('(')[-1].split(',')[0].strip())
```

```python
samsung
```

```python
samsung.drop(columns = ['Model', 'Number of rating and reviews', 'Battery', 'Display'], inplace = True)
```

```python
samsung
```

```python
samsung[~samsung['ram_rom'].str.contains('RAM')]
```

```python
samsung[~samsung['ram_rom'].str.contains('ROM')]
```

```python
samsung['ROM'] = samsung['ram_rom'].apply(lambda x: x.split('ROM')[0].split('|')[-1])
```

```python
samsung
```

```python
samsung['RAM'] = samsung['ram_rom'].apply(lambda x: x.split('RAM')[0].strip() if 'RAM' in x else '0')
```

```python
samsung
```

```python
samsung.drop(columns = ['ram_rom'], inplace = True)
```

```python
samsung.info()
```
