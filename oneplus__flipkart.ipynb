```python
# Setting up the ChromeDriver path
service = Service("C:/Users/user/Desktop/chromedriver.exe")
driver = webdriver.Chrome(service=service)

# Lists to store the data
model = []
ratings_reviews= []
discounted_price= []
discount= []
rating = []
camera = []
battery = []
ram_rom = []
display = []
original_price = []
product_image = []

# Loop to extract data from 10 pages
for j in range(1, 11):
    url = 'https://www.flipkart.com/q/oneplus-5g-phones?page={}'.format(j)
    
    # Open the URL with the WebDriver
    driver.get(url)
    time.sleep(5)  # Wait for the page to load
    
    # Get the page source and parse it with BeautifulSoup
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    soup.prettify()
    products = soup.find_all('div', class_='tUxRFH')
    
    for i in products:
        try:
            model.append(i.find('div', class_= 'KzDlHZ').text.strip())
        except:
            model.append(np.nan)
        try:
            ratings_reviews.append(i.find('span', class_="Wphh3N").text.strip())
        except:
            ratings_reviews.append(np.nan)
        try:   
            discounted_price.append(i.find('div', class_="Nx9bqj _4b5DiR").text.strip())
        except:
            discounted_price.append(np.nan)
        try:   
            original_price.append(i.find('div', class_="yRaY8j ZYYwLA").text.strip())
        except:
            original_price.append(np.nan)
        try:
            discount.append(i.find('div', class_= 'UkUFwK').text.strip())
        except:
            discount.append(np.nan)
        try:
            rating.append(i.find('div', class_="XQDdHH").text.strip())
        except:
            rating.append(np.nan)
        try:
            ram_rom.append(i.find_all('li', class_="J+igdf")[0].text.strip())
        except:
            ram_rom.append(np.nan)
        try:
            display.append(i.find_all('li', class_ = 'J+igdf')[1].text.strip())
        except:
            display.append(np.nan)
        try:
            camera.append(i.find_all('li', class_="J+igdf")[2].text.strip())
        except:
            camera.append(np.nan)
        try:
            battery.append(i.find_all('li', class_="J+igdf")[3].text.strip())
        except:
            battery.append(np.nan)
        try:
            product_image.append(i.find('img', class_="DByuf4")['src'])
        except:
            product_image.append(np.nan)

# Closing the WebDriver
driver.quit()

# Convering the fetched details into a DataFrame
d = {'Model': model, 'Number of rating and reviews': ratings_reviews, 'original_price': original_price, 'Discounted_Price': discounted_price, 'Discount': discount, 'Rating': rating, 'RAM_ROM': ram_rom, 'Battery': battery, 'Display': display, 'Camera': camera, 'product_image':  product_image}
df = pd.DataFrame(d)
print(df)
```

# Data Cleaning

```python
#Importing our csv file
oneplus = pd.read_csv('C:/Users/user/Downloads/one_plus_flipkart.csv')
```

```python
oneplus
```

```python
#Filling null values in discount column
oneplus['Discount'] = oneplus['Discount'].fillna('0% off')
```

```python
#Filling original price with discounted pricewhen discount is 0
for i in range(oneplus.shape[0]):
    if pd.isna(oneplus['original_price'].iloc[i]):
        oneplus['original_price'].iloc[i] = oneplus['Discounted_Price'].iloc[i]
```

```python
oneplus.dropna(subset=['Discounted_Price'], inplace = True)
```

```python
oneplus['Discounted_Price'] = oneplus['Discounted_Price'].str.replace('₹', '').str.replace(',', '').astype(int)
```

```python
oneplus['original_price'] = oneplus['original_price'].str.replace('₹', '').str.replace(',', '').astype(int)
```

```python
oneplus['Discount'] = oneplus['Discount'].apply(lambda x: x.split('%')[0]).astype(int)
```

```python
oneplus
```

```python
oneplus.info()
```

```python
oneplus['No_of_Ratings'] = oneplus['Number of rating and reviews'].astype(str).apply(lambda x: x.split(' ')[0])
```

```python
oneplus['No_of_Reviews'] = oneplus['Number of rating and reviews'].astype(str).apply(lambda x: x.split('&')[1].strip().split(' ')[0] if '&' in x else np.nan)
```

```python
oneplus['No_of_Reviews'] = oneplus['No_of_Reviews'].astype(str).apply(lambda x: x.split(' ')[0])
```

```python
oneplus['No_of_Ratings'] = oneplus['No_of_Ratings'].str.replace(',', '')
oneplus['No_of_Reviews'] = oneplus['No_of_Reviews'].str.replace(',', '')
```
